```python
import ast
```

```python
df['genres'][0]
```

```python
def get_genres(data):
  l = []
  for x in ast.literal_eval(data):
    l.append(x['name'])
  return l
```

```python
df['genres2'] = df['genres'].apply(get_genres)
```

```python
df
```

```python
df.drop(['genres'], axis = 1, inplace = True)
```

```python
df['cast'][0]
```

```python
def get_cast(data):
  l = []
  k = 0
  for x in ast.literal_eval(data):
    if(k<3):
      l.append(x['name'])
    else:
      break
    k+=1
  return l
```

```python
get_cast(df['cast'][0])
```

```python
df['cast2'] = df['cast'].apply(get_cast)
```

```python
df.drop(['cast'], axis = 1, inplace = True)
```

```python
df['crew'][0]
```

```python
def fetch_dir(data):
  l = []
  for x in ast.literal_eval(data):
    if(x['job'] == 'Director'):
      l.append(x['name'])
      break
  return l
```

```python
fetch_dir(df['crew'][0])
```

```python
df['director'] = df['crew'].apply(fetch_dir)
```

```python
df.drop(['crew'], axis = 1, inplace = True)
```

```python
df['keywords'][0]
```

```python
def fetch_keywords(data):
  l = []
  for x in ast.literal_eval(data):
    l.append(x['name'])
  return l
```

```python
df['keywords2']= df['keywords'].apply(fetch_keywords)
```

```python
df.drop(['keywords'], axis = 1, inplace = True)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem_text(data):
  l = []
  for x in data.split():
    l.append(ps.stem(x))
  return " ".join(l)
```
